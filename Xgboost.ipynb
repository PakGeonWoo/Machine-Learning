```python
import numpy as np
import pandas as pd # 데이터프레임을 위한 판다스
from sklearn.model_selection import train_test_split # 데이터 나누기
import xgboost as xgb # 모델 xgboost

######## 모델 저장 후 불러와 쓸 때 위 임폴트 작업을 해주어야한다! ########

# 성능 측정 지표
from sklearn.metrics import mean_absolute_error # MAE
from sklearn.metrics import mean_squared_error # MSE, RMSE
from sklearn.metrics import mean_squared_log_error # MSLE

# 시각화
import matplotlib
import matplotlib.pyplot as plt

# 모델 저장
import joblib
```

### 데이터 불러오기

```python
X_data = pd.read_csv('C:/Users/DELL/Desktop/X_data.csv', encoding='latin_1')
y_data = pd.read_csv('C:/Users/DELL/Desktop/y_data.csv', encoding='latin_1')
```

### 데이터 나누기 

```python
X = X_data.astype('float')
y = y_data.astype('float')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=6)
```

```python
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```

### 모델 선정 

```python
model = xgb.XGBRegressor(learning_rate=0.1,max_depth=7,n_estimators=100)
```

### 모델 학습 

```python
model.fit(X_train, y_train, verbose=False)
```

### 모델 평가

```python
model.score(X_train, y_train)
```

```python
preds = model.predict(X_test)
```

```python
print("MAE(평균절대오차) >> ", mean_absolute_error(y_test, preds)) # MAE(평균절대오차)
print("MSE(평균제곱오차) >> ", mean_squared_error(y_test, preds)) # MSE(평균제곱오차)
MSE = mean_squared_error(y_test, preds) 
print("RMSE(평균제곱근오차) >> ", np.sqrt(MSE)) # RMSE(평균제곱근오차)
print("MSLE(MSE방식에 로그를 적용) >> ", mean_squared_log_error(y_test, preds)) #MSLE(MSE방식에 로그를 적용)

print("")
def MAPE(y_test, preds): # MAPE(평균절대비오차)
    return np.mean(np.abs((y_test - preds) / y_test)) * 100 
print("MAPE(평균절대비오차)\n", MAPE(y_test, preds))

print("")
def MPE(y_test, preds): # MPE(평균백분율오차)
    return np.mean((y_test - preds) / y_test) * 100
print("MPE(평균백분율오차)\n", MPE(y_test, preds))
```

### 시각화
